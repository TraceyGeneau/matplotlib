It was determined that of the 10 treatments, the four with the smallest mean final tumor volume (mm3) were Capomulin, Ramicane, Infubinol and Ceftamin. There was very little difference in the mean of Capomulin and Ramicane wheras Infubinol and Ceftamin appeard to have higher final tumor volume (mm3) means.  Statistical analysis would have to be carried out to determine if there was a signicant differnce between the first two drug regimens and the last two.  In additon to this, the mean of infubinol may be skewed slightly lower as there was an outlier observed at 36.2 (mm3).

From the linear graph of Mouse l509's tumor volume (mm3) over time, it could be observed that a pattern in a slight increase in volume size followe by a bigger drop was present.  It would be interesting to continue the trial pass 45 days to see if the trend continued as well as look at othter drug treatments to see if a similar pattern was observed.  If not, capomulin may show an inflated tumor size as 45 days compared to other treatment plans at this timepoint or even if this is an isolated event that occured with moust l509.

When the final tumor volume for Capomulin was compared to the SCC tumor weight, it was observed that a positive correlation was present (0.84).  This would further back up the findings for mouse l509 indicating that there was a pattern in growth of the tumor followed by a decrease in tumor size. 

Overall more statistical analysis would need to be carried out to get a better understanding of the interaction of the drug regimens with the SCC Tumor cell volume (mm3) in mice.  From simplly comparing the means of the final tumor volumes, Capomulin and Ramicane appeared to result in the smallerst growth due to their smaller mean final tumor volumes with Cetamin and Infubinol following close behind.  

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
study_complete = pd.merge(study_results,mouse_metadata, how="left", on=["Mouse ID","Mouse ID"])
study_complete.head()

# Display the data table for preview
```

```python
# Checking the number of mice.
unique = study_complete["Mouse ID"].unique()
mouse_count = study_complete["Mouse ID"].nunique()
print(f'The total number of mice in the study is {mouse_count}.')
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 

#Reference for duplicates https://sparkbyexamples.com/pandas/pandas-get-list-of-all-duplicate-rows/#:~:text=Pandas%20DataFrame.,multiple%20columns%20or%20all%20columns.

duplicates = study_complete[study_complete[["Mouse ID", "Timepoint"]].duplicated(keep=False)==True] 
duplicates

#Notes for me
#reference 
#keep{‘first’, ‘last’, False}, default ‘first’
#Determines which duplicates (if any) to keep.
#‘first’ : Drop duplicates except for the first occurrence.
#‘last’ : Drop duplicates except for the last occurrence.
#False : Drop all duplicates.
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.

#make a data frame that does not include g989 as we saw above it has duplicates
clean_df = study_complete[study_complete["Mouse ID"]!="g989"]
clean_df.head()
```

# Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.
```

```python
#Mean Tumor Vol for each Drug Regimen
mean_tumor = clean_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].mean()
mean_tumor
```

```python
#Median Tumor vol for each Drug Regimen
median_tumor = clean_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].median()
median_tumor

```

```python
#Variance in Tumor Volume
var_tumor = clean_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].var()
var_tumor
```

```python
#Standard Deviation of Tumor volume
std_tumor = clean_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].std()
std_tumor
```

```python
#Standard Error of Tumor Volume
sem_tumor = clean_df.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].sem()
sem_tumor
```

```python
stat_summary = pd.DataFrame({"Mean Tumor Volume":mean_tumor,
                            "Median Tumor Voulme":median_tumor,
                            "Variance Tumor Volume":var_tumor,
                            "Tumor Volume Standard Deviation":std_tumor,
                            "Tumor Volume Standard Error":sem_tumor})

stat_summary
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
#Reference for this https://sparkbyexamples.com/pandas/pandas-aggregate-functions-with-examples/

summary_stat = clean_df.groupby("Drug Regimen").aggregate({"Tumor Volume (mm3)":['mean','median','var','std','sem']})
summary_stat
```

# Quartiles, Outliers and Boxplots

```python
#we will use the Clean_df database

clean_df.head()
```

```python
#Remove Columns that we do not need

tumor_vol_df = clean_df[["Mouse ID","Timepoint","Tumor Volume (mm3)","Drug Regimen"]]
tumor_vol_df.head()

```

```python
#sorting the columns by mouse ID and timepoint.  set the index to follow

tumor_sorted_mouse = tumor_vol_df.sort_values(["Mouse ID",'Timepoint'],
                    ascending=True)
tumor_sorted_mouse=tumor_sorted_mouse.reset_index(drop=True)
tumor_sorted_mouse.head(15)

```

```python
#isolate the four best Drug Regimens Campomulin, Ramicane, Infubinol and Ceftamin

iso_treatment_df = tumor_sorted_mouse.loc[(tumor_sorted_mouse["Drug Regimen"]=="Capomulin")| 
                                          (tumor_sorted_mouse["Drug Regimen"]=="Ramicane")| 
                                          (tumor_sorted_mouse["Drug Regimen"]=="Infubinol")| 
                                          (tumor_sorted_mouse["Drug Regimen"]=="Ceftamin"),
                               ["Mouse ID","Timepoint","Tumor Volume (mm3)","Drug Regimen"]]
iso_treatment_df.head()

```

```python
#go through and drop all of the duplicate mouse ID but keep the last one of each group.
#this will give the final timepoint and final volume

final_df = iso_treatment_df.drop_duplicates(subset='Mouse ID', keep="last")
final_df

#notice 100 rows
```

```python
#Rename the Tumor volume to final volume
final_df = final_df.rename(columns={"Tumor Volume (mm3)":"Final Tumor Vol (mm3)","Timepoint":"Final Timepoint"})
final_df.head()
```

```python
#Pull initial Cell volumes per mouse Id

initial_df = iso_treatment_df.drop_duplicates(subset='Mouse ID', keep="first")
initial_df   

#notice 100 rows
```

```python
#Rename the Tumor volume to initial volume
initial_df = initial_df.rename(columns={"Tumor Volume (mm3)":"Initial Tumor Vol (mm3)","Timepoint":"Initial Timepoint"})
initial_df.head()
```

```python
# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
#Reference for the joining of two columns https://sparkbyexamples.com/pandas/pandas-merge-two-dataframes-on-multiple-columns/

tumor_vol_df = pd.merge(initial_df,final_df, how="left", left_on=["Mouse ID","Drug Regimen"], right_on=["Mouse ID","Drug Regimen"])
tumor_vol_df.head(15)

#The directions say to calculate the total final volume for each mouse but
#the final volume was measured.  There should not be a calculation as subtracting
#would give the change in volume not the final volume. 

```

```python
# Put treatments into a list for for loop

treatment_list=["Capomulin","Ramicane","Infubinol","Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
final_tumor_vol_ls=[]

#loop through each drug in the treament list, locating the rows in the merged DataFrame
#that correspond to each treatment.  Append to resulting final tumor volumes for each drug
#put the data into the list in quartiles for graphing


for drug in treatment_list:
    tumor_vol = tumor_vol_df.loc[tumor_vol_df["Drug Regimen"] == drug, 'Final Tumor Vol (mm3)']
    final_tumor_vol_ls.append(tumor_vol)
    
    quartiles = tumor_vol.quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    outliers = tumor_vol.loc[(tumor_vol < lower_bound) | (tumor_vol > upper_bound)]
    print(f"{drug}'s potential outliers: {outliers}")
    

```

```python
#generate box plot

mark_out = dict(markerfacecolor='red',markersize=12)
plt.boxplot(final_tumor_vol_ls, labels = treatment_list,flierprops=mark_out)
plt.ylabel('Final Tumor Volume (mm3)')
plt.show()
```

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin

mouse_l509 = clean_df.loc[clean_df['Mouse ID'] == "l509"]
mouse_data = mouse_l509.groupby(['Timepoint'])[['Tumor Volume (mm3)']].mean()
mouse_data
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen

capomulin_table = clean_df.loc[clean_df['Drug Regimen'] == "Capomulin"]
capomulin_average = capomulin_table.groupby(['Mouse ID'])[['Weight (g)', 'Tumor Volume (mm3)']].mean()
capomulin_average.head()
```

```python
#Scatter Plot
plt.scatter(capomulin_average['Weight (g)'],capomulin_average['Tumor Volume (mm3)'])
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.show()
```

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen


corr=round(st.pearsonr(capomulin_average['Weight (g)'],capomulin_average['Tumor Volume (mm3)'])[0],2)
print(f"The correlation between mouse weight and the average tumor volume is {corr}")

#model lingress (x,y=none)
model = st.linregress(capomulin_average['Weight (g)'],capomulin_average['Tumor Volume (mm3)'])
y_values = capomulin_average['Weight (g)']*model[0]+model[1]

#plot scatter plot
plt.scatter(capomulin_average['Weight (g)'],capomulin_average['Tumor Volume (mm3)'])

#Print line of regression
plt.plot(capomulin_average['Weight (g)'],y_values,color="red")

#Print Graph
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.show()

```
